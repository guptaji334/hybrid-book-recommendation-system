
## Popularity Based Recommendation System


```python
#Data Frame for Books
books.head()
```

```python
#DataFrame for Ratings
ratings.head()
```

```python
#DataFrame for Users
users.head()
```

```python
# Joining books and user ratings into one table
books_with_ratings = ratings.merge(books, on = 'ISBN') #using ISBN as reference for merging
```

```python
books_with_ratings.head()
```

```python
books_with_ratings.shape
```

```python
popular_df = books_with_ratings.groupby('Book-Title').agg(num_rating = ('Book-Rating', 'count'),
                                                          avg_rating = ('Book-Rating', 'mean'))
```

```python
popular_df = popular_df.reset_index() #adds indexes to the popularity dataframe
```

```python
popular_df 
```

```python
popular_df.sort_values('num_rating',ascending=False)
```

```python
#Popularity is based on the no. of people read the book (taking ' num_rating'> 300)
#It is based on the rating it got.
popular_df = popular_df[popular_df['num_rating']>300].sort_values('avg_rating', ascending = False)
```

```python
popular_df
```

```python
popular_df = popular_df.head(50)
```

```python
popular_df
```

```python
# For the model deployment I need book-title, author, image Url 
popular_df = popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title',
                                                                                       'Book-Author',
                                                                                       'Image-URL-M',
                                                                                       'num_rating',
                                                                                       'avg_rating']]
```

### Collaborative Filtering
- Similar Book Prediction based on users feedback

```python
books_with_ratings.head()
```

```python
# Grouping based on user-id tells the no. of books rated by each user:
x = books_with_ratings.groupby("User-ID").count()
```

```python
x
```

```python
x.index
```

```python
x.shape
```

```python
# applying criteria to reduce data redundancy
# Select only users who atleast gave feedback for 200 books[Power Users]
x = x['Book-Rating'] > 200
```

```python
x
```

```python
# ABove are persons who haven't read 200 books atleast
power_users = x[x].index
```

```python
power_users.shape
```

```python
power_users.sort_values()
```

```python
 #selecting only records of power users
filtered_ratings = books_with_ratings[books_with_ratings['User-ID'].isin(power_users)]
```

```python
filtered_ratings
```

```python
# considering only the best users (atleast 200 books feedback) group them based on the book title.
y = filtered_ratings.groupby('Book-Title').count()
```

```python
y
```

```python
# The above dataframe tells how many users have read the book.
```

```python
y.sort_values('User-ID', ascending = False)
```

```python
y = y['User-ID'] >= 50
```

```python
y
```

```python
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
#pivot table giving the ratings for each books from each user
# Book row with user_id as column
pt = final_ratings.pivot_table(index='Book-Title', columns ='User-ID', values = 'Book-Rating')
```

```python
pt
```

```python
pt = pt.fillna(0)
```

```python
pt
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
type(similarity_scores)
```

```python
 df_temp = pd.DataFrame(similarity_scores)
```

```python
df_temp
```

```python
pt.index
```

```python
df_temp_name = df_temp
```

```python
df_temp_name.index = pt.index
df_temp_name.columns = pt.index
```

```python
df_temp_name
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similarity_items = sorted(list(enumerate(similarity_scores[index])),key= lambda x:x[1], reverse= True)[1:6]
    # creating empty list and in that list populating with book information
    #book author book-title image-URL
    #empty list
    data = []
    for i in similarity_items:
        item=[]
        temp_df = books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```
