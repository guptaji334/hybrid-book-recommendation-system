
## Popularity Based Recommendation System


```python
#Data Frame for Books
books.head()
```

```python
#DataFrame for Ratings
ratings.head()
```

```python
#DataFrame for Users
users.head()
```

```python
# Joining books and user ratings into one table
books_with_ratings = ratings.merge(books, on = 'ISBN') #using ISBN as reference for merging
```

```python
books_with_ratings.head()
```

```python
books_with_ratings.shape
```

```python
popular_df = books_with_ratings.groupby('Book-Title').agg(num_rating = ('Book-Rating', 'count'),
                                                          avg_rating = ('Book-Rating', 'mean'))
```

```python
popular_df = popular_df.reset_index() #adds indexes to the popularity dataframe
```

```python
popular_df 
```

```python
popular_df.sort_values('num_rating',ascending=False)
```

```python
#Popularity is based on the no. of people read the book (taking ' num_rating'> 300)
#It is based on the rating it got.
popular_df = popular_df[popular_df['num_rating']>300].sort_values('avg_rating', ascending = False)
```

```python
popular_df
```

```python
popular_df = popular_df.head(50)
```

```python
popular_df
```

```python
# For the model deployment I need book-title, author, image Url 
popular_df = popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title',
                                                                                       'Book-Author',
                                                                                       'Image-URL-M',
                                                                                       'num_rating',
                                                                                       'avg_rating']]
```

```python
popular_df
```

### Colaborative Filtering
- Similar Book Prediction based on users feedback

```python
books_with_ratings.head()
```

```python
# Grouping based on user-id tells the no. of books rated by each user:
x = books_with_ratings.groupby("User-ID").count()
```

```python
x
```

```python
x.index
```

```python
x.shape
```

```python
# applying criteria to reduce data redundancy
# Select only users who atleast gave feedback for 200 books[Power Users]
x = x['Book-Rating'] > 200
```

```python
x
```

```python
# Below indexes shows  persons who have read 200 books atleast
power_users = x[x].index
```

```python
power_users
```

```python
power_users.sort_values()
```

```python
#selecting only records of power users
filtered_ratings = books_with_ratings[books_with_ratings['User-ID'].isin(power_users)]
```

```python
filtered_ratings
```

```python
filtered_ratings.shape
```

```python
# Considering only best users (atleast 200 books feedback) group them based on book title.
y = filtered_ratings.groupby('Book-Title').count()
```

```python
y
```

```python
# the above dataframe tells how many users have read the book.
y.sort_values('User-ID',ascending = False)
```

```python
y = y['User-ID'] >= 50
```

```python
y
```

```python
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```
